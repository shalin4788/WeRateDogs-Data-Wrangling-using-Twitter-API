```python
# Create dataframe by reading twitter archive enhanced dataset that has tweet details in csv format
df_1 = pd.read_csv('C:\\Shalin\\Data science\\DAND folders and certification\\projects\\Project 4 - Data wrangling - Twitter API\\twitter-archive-enhanced.csv')
```

```python
# Extract the tweet_id values from the dataframe generated
tweet_ids = df_1.tweet_id.values
len(tweet_ids)
```

```python
# Generate tweet_json text file by extracting tweet_id and relevant data associated to  tweet_id in the dataframe
count = 0
fails_dict = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('C:\\Shalin\Data science\\DAND folders and certification\\projects\\Project 4 - Data wrangling - Twitter API\\tweet_json.txt', 'w') as outfile:
    # This loop will likely take 20-30 minutes to run because of Twitter's rate limit
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

```python
# Extract tweet id, retweet count and favorite count columns from the tweet_json file generated
twitter_list = pd.DataFrame(columns = ['tweet_id', 'retweet_count', 'favorite_count','retweeted','created_at'])

with open('C:\\Shalin\\Data science\\DAND folders and certification\\projects\\Project 4 - Data wrangling - Twitter API\\tweet_json.txt') as f:
    for line in f:
        status = json.loads(line)
        tweet_id = status['id_str']
        #print(tweet_id)
        retweet_count = status['retweet_count']
        #print(retweet_count)
        favorite_count = status['favorite_count']
        #print(favorite_count)
        created_at = status['created_at']
        #print(created_at)
        retweeted = status['retweeted']
        twitter_list = twitter_list.append(pd.DataFrame([[tweet_id, retweet_count, favorite_count, retweeted, created_at]],
                                                       columns = ['tweet_id', 'retweet_count', 'favorite_count','retweeted','created_at']))

twitter_list = twitter_list.reset_index(drop=True)
twitter_list
```

```python
# Download tweet image predictions from image_predictions.tsv file programmatically using Request library by creating a dataframe
df_2 = pd.read_csv('C:\\Shalin\\Data science\\DAND folders and certification\\projects\\Project 4 - Data wrangling - Twitter API\\image-predictions.tsv', sep = '\t')
```

```python
# Merge df1 and df2 to create a new dataframe with image predictions for dog dataset based on tweet_id along with retweet count and favorite count
df_combined = pd.concat([twitter_list, df_2], axis=1, join='outer')
df_combined
```
